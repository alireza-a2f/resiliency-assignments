```python
!pip install 'networkx==2.6.3'
import csv
import networkx as nx
import matplotlib.pyplot as plt

def read_edge_data(filename):
    edge_list = []
    with open(filename) as csvfile:
        data = csv.reader(csvfile)
        next(data, None)  # Skip header
        for row in data:
            edge_list.append((row[0], row[1]))
    return edge_list

file_name = 'graph.csv'
edge_data = read_edge_data(file_name)

G = nx.Graph(edge_data)

print('Adjacency Matrix:')
print(nx.adjacency_matrix(G))

N = len(G.nodes)
L = len(G.edges)
S = nx.number_connected_components(G)

print('Number of Nodes:', N)
print('Number of Edges:', L)
print('Number of Connected Components:', S)
print('mu:', L - N + S)
print('alpha_p:', (L - N + S) / (2 * N - 6 + S))
print('beta:', L / N)
print('gama_p:', L / (3 * N - 6))
print('Network Diameter:', nx.diameter(G))
print('Degree:', {node: G.degree(node) for node in G.nodes})
print('Closeness Centrality:', {node: nx.closeness_centrality(G, node) for node in G.nodes})
print('Betweenness Centrality:', nx.betweenness_centrality(G))

print('Simple')
nx.draw(G, with_labels=True, node_color='lightblue')
plt.show()

print('Circular')
nx.draw_circular(G, with_labels=True, node_color='lightblue')
plt.show()

print('Kamada Kawai')
nx.draw_kamada_kawai(G, with_labels=True, node_color='lightblue')
plt.show()

print('Planar')
nx.draw_planar(G, with_labels=True, node_color='lightblue')
plt.show()

print('Spectral')
nx.draw_spectral(G, with_labels=True, node_color='lightblue')
plt.show()

print('Spring')
nx.draw_spring(G, with_labels=True, node_color='lightblue')
plt.show()

print('Shell')
nx.draw_shell(G, with_labels=True, node_color='lightblue')
plt.show()
```
